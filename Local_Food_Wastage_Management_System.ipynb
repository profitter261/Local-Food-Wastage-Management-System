## 8) How many food claims have been made for each food item?

```python
import pandas as pd
import sqlite3

# Load your CSVs
foodlisting = pd.read_csv('/content/food_listings_data.csv')
claims = pd.read_csv('/content/claims_data.csv')

# Create in-memory SQLite database
conn = sqlite3.connect(":memory:")

# Upload both DataFrames to SQLite tables
claims.to_sql('claims', conn, index=False, if_exists='replace')
foodlisting.to_sql('foodlisting', conn, index=False, if_exists='replace')

# JOIN query to combine employee with their department names
query = """
SELECT
    t3.Food_Name,
    count(t2.Claim_ID) no_food_claims
FROM
    foodlisting t3
JOIN
    claims t2
ON
    t2.Food_ID = t3.Food_ID
group by t3.Food_Name
order by no_food_claims desc
"""

# Run query
result11 = pd.read_sql_query(query, conn)
result11
```

##11) Which City Has Fastest Claim Rate

```python
import pandas as pd
import sqlite3

# Load datasets
foodlisting = pd.read_csv('/content/food_listings_data.csv')
claims = pd.read_csv('/content/claims_data.csv')

# Convert date columns to datetime
foodlisting['Expiry_Date'] = pd.to_datetime(foodlisting['Expiry_Date'], errors='coerce')
claims['Timestamp'] = pd.to_datetime(claims['Timestamp'], errors='coerce')

# Drop rows with invalid dates
foodlisting = foodlisting.dropna(subset=['Expiry_Date'])
claims = claims.dropna(subset=['Timestamp'])

# Create in-memory SQLite DB
conn = sqlite3.connect(":memory:")
claims.to_sql('claims', conn, index=False, if_exists='replace')
foodlisting.to_sql('foodlisting', conn, index=False, if_exists='replace')

# Query with capped percentage at 0
query = """
SELECT
    f.Location AS city,
    ROUND(MAX(AVG(julianday(c.Timestamp) - julianday(f.Expiry_Date)) * 100.0 / 30, 0), 2) AS avg_claim_time_percent
FROM
    foodlisting f
JOIN
    claims c ON f.Food_ID = c.Food_ID
WHERE
    c.Status = 'Completed'
GROUP BY
    city
ORDER BY
    avg_claim_time_percent DESC
"""

# Execute the query
result12 = pd.read_sql_query(query, conn)
result12
```

##3) At which time the Food demand from recievers is increased?

```python
foodlisting = pd.read_csv('/content/food_listings_data.csv')
claims = pd.read_csv('/content/claims_data.csv')

# Create in-memory SQLite database
conn = sqlite3.connect(":memory:")

# Upload both DataFrames to SQLite tables
claims.to_sql('claims', conn, index=False, if_exists='replace')
foodlisting.to_sql('foodlisting', conn, index=False, if_exists='replace')

query = '''
SELECT
    f.Meal_Type,
    CAST(SUM(CASE WHEN c.Status = 'Completed' THEN 1 ELSE 0 END) AS REAL) * 100 / (SELECT COUNT(*) FROM claims c where Status = 'Completed') as success_percentage
FROM
    claims c
JOIN
    foodlisting f ON c.Food_ID = f.Food_ID
GROUP BY
    f.Meal_Type
ORDER BY success_percentage DESC;
'''

# Run query
result19 = pd.read_sql_query(query, conn)
result19
```
